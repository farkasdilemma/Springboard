**Cluster Analysis**

```python
df.groupby('Clusters').mean()
```

Tracks in all three clusters fall relatively closely in duration

```python
_=df.groupby('Clusters')['duration_ms'].hist()
```

**Loudness had distinct classes with cluster 0 being the quietest and cluster 1 being the loudest on average**

```python
df.groupby('Clusters')['loudness'].describe()
```

```python
_=df.groupby('Clusters')['loudness'].hist()
```

**Valence is spotify's coding for how happy a track feels with anything above 0.5 tending to sound more happy while anything below sounding more sad. Cluster 1 is much higher/happier than the other clusters**

```python
df.groupby('Clusters')['valence'].describe()
```

**Energy is relatively high for clusters 1 and 2 compared to cluster 0**

```python
df.groupby('Clusters')['energy'].describe()
```

**Acousticness is how 'live' a track feels. Values close to 1 are more likely to be live records. By Comparison Cluster 0 is the most likely to be acoustic recordings**

```python
df.groupby('Clusters')['acousticness'].describe()
```

**Mode is Spotify's coding for whether the track is in major (1) or minor (0) key. This is somewhat confirmed by the valence scores. Since minor keys tend to sound more sad, it makes sense that a larger percentage of songs in cluster 2 is in the minor key (98/230) and it also has the lowest valence**

```python
df.groupby('Clusters')['mode'].value_counts()
```

Generally, the clusters appear to breakdown tracks into three buckets. Cluster 1 is the upbeat, happy playlist with the highest valence, loudness, energy and danceability. And while both clusters 0 and 2 tend to be sad sounding, cluster 0 is also much more likely to be an acoustic track, with less energy and also much quieter by comparison to the others. This is suggests a quieter, live performance compared to a lively sounding studio recording.
