```python
from sklearn.preprocessing import MinMaxScaler

min_max_scaler = MinMaxScaler()

df_x = df.drop(['album name','track name','artist_0']
                  , axis =1)

minmax_x = min_max_scaler.fit_transform(df_x)
```

```python
fig = plt.figure(figsize=(25,10)) 
_=pd.DataFrame(dict(zip(df_x.columns,minmax_x))).boxplot()
```

**In order for KMeans to be as effective as possible, all features have been standardized and weighs equally in the algorithm. Additionally, since this is an unsupervised learning technique, there will be no training and testing dataset splitting. Instead, the performance will be directly evaluated based on whether or not the recommended tracks make sense for a human listener in the next step.**
